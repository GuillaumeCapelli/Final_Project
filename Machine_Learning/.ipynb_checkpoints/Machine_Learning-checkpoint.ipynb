```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Loading the Jigsaw challenge training data
train_file_path = 'train_clean.csv'
train_df = pd.read_csv(train_file_path)
```

```python
# Create an additional category for comments with no toxic labels
train_df['non_toxic'] = (train_df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].sum(axis=1) == 0).astype(int)
```

```python
train_df
```

```python
# Calculate the sum of each label type again, including the 'non_toxic' category
label_sums_with_non_toxic = train_df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate', 'non_toxic']].sum().sort_values(ascending=False)
```

```python
# Create a bar plot for the label frequency including 'non_toxic' category
plt.figure(figsize=(10, 6))
bar_plot = label_sums_with_non_toxic.plot(kind='bar', color='skyblue')
plt.title('Frequency of Each Label Type in the Training Data (Including Non-Toxic Comments)')
plt.xlabel('Label Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.7)

# Add labels on top of the bars
for bar in bar_plot.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

plt.show()
```

```python
# Preprocessing: Vectorizing the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X = tfidf_vectorizer.fit_transform(train_df['comment'])
y = train_df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']]
```

```python
X
```

```python
y
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# Training Logistic Regression Model
logreg_model = OneVsRestClassifier(LogisticRegression(solver='liblinear'))
logreg_model.fit(X_train, y_train)

# Training SVM Model
svm_model = OneVsRestClassifier(LinearSVC(random_state=123))
svm_model.fit(X_train, y_train)
```

```python
import pickle

# Serializing the logistic regression model
with open('logreg_model.pkl', 'wb') as file:
    pickle.dump(logreg_model, file)

# Serializing the SVM model
with open('svm_model.pkl', 'wb') as file:
    pickle.dump(svm_model, file)

# Serializing the TF-IDF vectorizer
with open('tfidf_vectorizer.pkl', 'wb') as file:
    pickle.dump(tfidf_vectorizer, file)
```

```python
from sklearn.metrics import classification_report
import pandas as pd

# Function to evaluate the model and return the report as a DataFrame
def evaluate_model_and_return_df(model, X_test, y_test):
    y_pred = model.predict(X_test)
    report_dict = classification_report(y_test, y_pred, target_names=y.columns, zero_division=0, output_dict=True)
    report_df = pd.DataFrame(report_dict).transpose()
    accuracy = accuracy_score(y_test, y_pred)
    return report_df, accuracy

# Evaluating the models and obtaining the reports as DataFrames
report_logreg_df, accuracy_logreg = evaluate_model_and_return_df(logreg_model, X_test, y_test)
report_svm_df, accuracy_svm = evaluate_model_and_return_df(svm_model, X_test, y_test)
```

```python
# Displaying the reports
print("\nClassification report for the logistic regression model:")
print(report_logreg_df)
print("Accuracy for the logistic regression model:", accuracy_logreg)

print()
print()

print("\nClassification report for the SVM model:")
print(report_svm_df)
print("Accuracy for the SVM model:", accuracy_svm)
```

```python
# Function to find the first index of a comment matching the label criteria
def find_first_index_for_labels(target_labels):
    indices = new_data_pred_logreg_df.index[new_data_pred_logreg_df.eq(target_labels).all(axis=1)]
    return indices[0] if len(indices) > 0 else None

# Set option to display full (non-truncated) comment text
pd.set_option('display.max_colwidth', None)

# Finding the first index for each scenario
non_toxic_index = find_first_index_for_labels(pd.Series([0] * len(y.columns), index=y.columns))
toxic_index = find_first_index_for_labels(pd.Series([1, 0, 0, 0, 0, 0], index=y.columns))
obscene_index = find_first_index_for_labels(pd.Series([0, 0, 1, 0, 0, 0], index=y.columns))
toxic_obscene_index = find_first_index_for_labels(pd.Series([1, 0, 1, 0, 0, 0], index=y.columns))

# Collecting indices
selected_indices = [index for index in [non_toxic_index, toxic_index, obscene_index, toxic_obscene_index] if index is not None]

# Creating a DataFrame for the selected comments and their labels
selected_comments_df = new_data_df.loc[selected_indices]
selected_labels_df = new_data_pred_logreg_df.loc[selected_indices]
result_df = pd.concat([selected_comments_df.reset_index(drop=True), selected_labels_df.reset_index(drop=True)], axis=1)

result_df
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

# Define the grid parameters for the SVM with a special focus on 'max_iter'
param_grid_svm = {
    'estimator__C': [0.01, 0.1, 1],
    'estimator__max_iter': [1000, 5000, 10000],  # Increasing the number of iterations
    'estimator__class_weight': [None, 'balanced']
}

# Setting up GridSearchCV for the SVM model
grid_search_svm = GridSearchCV(
    estimator=OneVsRestClassifier(LinearSVC(dual=False, random_state=123)),  # dual=False for large datasets
    param_grid=param_grid_svm,
    scoring='f1_weighted',  # Using f1_weighted to account for class imbalance
    cv=5,
    verbose=3
)

# Training with grid search
grid_search_svm.fit(X_train, y_train)

# Displaying the best parameters and score
print("Best parameters for SVM:", grid_search_svm.best_params_)
print("Best score for SVM:", grid_search_svm.best_score_)

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the grid parameters for logistic regression
param_grid_logreg = {
    'estimator__C': [0.1, 1],  # Start with a smaller range
    'estimator__max_iter': [1000, 3000],  # Start with a lower max_iter
    # Removed solver from grid to use default 'lbfgs' which is good for multi-class problems
}

grid_search_logreg = GridSearchCV(
    estimator=OneVsRestClassifier(LogisticRegression()),
    param_grid=param_grid_logreg,
    scoring='f1_weighted',
    cv=5,
    verbose=3,
    n_jobs=-1  # Use all CPU cores
)

grid_search_logreg.fit(X_train, y_train)

# Displaying the best parameters and score
print("Best parameters for logistic regression:", grid_search_logreg.best_params_)
print("Best score for logistic regression:", grid_search_logreg.best_score_)
```

```python
import pickle

# Loading the logistic regression model
with open('logreg_model.pkl', 'rb') as file:
    loaded_logreg_model = pickle.load(file)

# Loading the SVM model
with open('svm_model.pkl', 'rb') as file:
    loaded_svm_model = pickle.load(file)

# Loading the TF-IDF vectorizer
with open('tfidf_vectorizer.pkl', 'rb') as file:
    loaded_tfidf_vectorizer = pickle.load(file)
```

```python
def classify_comment_into_df(tfidf_vectorizer, logreg_model, svm_model, categories):
    # Taking user input for the comment
    comment = input("Enter your comment: ")

    # Preprocessing the input comment using the same TF-IDF vectorizer
    processed_comment = tfidf_vectorizer.transform([comment])

    # Applying the trained models to the processed comment
    pred_logreg = logreg_model.predict(processed_comment)
    pred_svm = svm_model.predict(processed_comment)

    # Creating a DataFrame for the predictions
    data = []
    for category, pred_lr, pred_svm in zip(categories, pred_logreg[0], pred_svm[0]):
        data.append({'Category': category, 
                     'Logistic Regression': bool(pred_lr), 
                     'SVM': bool(pred_svm)})

    predictions_df = pd.DataFrame(data)
    return predictions_df

# Example usage of the function
categories = y.columns  # The categories of toxicity

predictions_df = classify_comment_into_df(loaded_tfidf_vectorizer, loaded_logreg_model, loaded_svm_model, categories)
predictions_df
```
