```python
import os
from google.cloud import bigquery
import pandas as pd
```

```python
# Set the environment variable to the location of your credentials file
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "C:\\Users\\Guillaume\\AppData\\Roaming\\gcloud\\application_default_credentials.json"
```

```python
# Create a BigQuery client
client = bigquery.Client()

# Define your query here - make sure you have the correct table name and dataset ID
query = """
SELECT text, hn.by as author, timestamp, type
FROM `bigquery-public-data.hacker_news.full` as hn
WHERE text IS NOT NULL AND hn.by IS NOT NULL AND timestamp IS NOT NULL
ORDER BY RAND()
LIMIT 3000
"""

# Run the query and convert the results to a dataframe
df = client.query(query).to_dataframe()
df
```

```python
# Save the dataframe to a CSV file
df.to_csv('hacker_news_sample.csv', index=False)

# Print the first few rows of the dataframe to confirm it looks as expected
df
```
